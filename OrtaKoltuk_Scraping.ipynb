```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url_link = ["https://izleryazar.com/category/sinema/film-yorumlari/"]
def url_generator(url,total_page_num):  #Sayfaların Url lerini oluşturur.
    arr = []
    for i in range(1,total_page_num+1):
        new_url = (f"{url}/page/{i}")
        arr.append(new_url)
    return arr
```

Toplam Sayfa Sayısını Alır

```python
url = "https://izleryazar.com/category/sinema/film-yorumlari/"
req = requests.get(url)
html = req.text
parsedHtml = BeautifulSoup(html,"html.parser")
total_page_num = parsedHtml.find("div" , {"class" :"page-nav td-pb-padding-side"})
total_page_num = int(total_page_num.find("a",{"class":"last"}).getText())
total_page_num
```

```python
page_url_arr = url_generator(url,total_page_num)
```

```python
print(len(page_url_arr))
```

Gelen URL lere göre tüm linkleri alma

```python
link_arr = []
def get_link_from_url(url):
    req = requests.get(url)
    html = req.text
    parsedHtml = BeautifulSoup(html,"html.parser")
    link_divs = parsedHtml.find("div",{"class":"td-pb-span8 td-main-content"}).find_all("div", {"class": "td-module-thumb"})

    for link_div in link_divs:
        lnk = link_div.find("a")
        if lnk and 'href' in lnk.attrs:
            href_value = lnk['href']
            link_arr.append(href_value)
```

```python
for i in range(1,len(page_url_arr)):
    get_link_from_url(page_url_arr[i])
```

```python
def get_content(url):
    req = requests.get(url)
    html = req.text
    parsedHtml = BeautifulSoup(html, "html.parser")
    title = parsedHtml.find("div", {"class": "td-post-content td-pb-padding-side"}).find("p").text
    soup = parsedHtml.find("div", {"class": "td-post-content td-pb-padding-side"}).find_all("p")
    film_content =""
    for p_tag in soup:
        film_content += p_tag.text.strip()
    film_content = film_content.replace("Ismimi, e-postamı ve web sitemi bir dahaki sefere bu tarayıcıya kaydet.Δ","")
    title =(f"{title} hakkında ne düşünüyorsun?")
    json_ = {
    "title": title,
    "content": film_content
    }
    total_content = []
    total_content.append({
            "title": title,
            "content": film_content
        })
    return total_content
```
