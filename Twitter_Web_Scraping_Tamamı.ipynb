```python
def get_tweet_by_id(tweet_id,res):
    api_url = f"https://publish.twitter.com/oembed?url=https://twitter.com/twitter/status/{tweet_id}"
    text =""
    res_list =""
    try:
        # HTTP GET isteƒüi g√∂nder
        response = requests.get(api_url)
        response.raise_for_status()  # Hata durumunda isteƒüi sonlandƒ±r

        # JSON yanƒ±tƒ±nƒ± al
        tweet_data = response.json()

        # HTML i√ßeriƒüini BeautifulSoup ile parse et
        soup = BeautifulSoup(tweet_data['html'], 'html.parser')

        # Tweet metnini √ßek
        tweet_text = soup.find('p').text

        # Eƒüer tweet_text i√ßinde "http" varsa, bu kƒ±smƒ± atla
        if "http" in tweet_text:
            http_index = tweet_text.find("http")
            tweet_text = tweet_text[:http_index].strip()
            print(tweet_text)

        # "Hata" ile ba≈ülamƒ±yorsa text listesine ekle
        if not tweet_text.startswith("Hata"):
            text = tweet_text
            res_list= res


    except requests.exceptions.RequestException as e:
        print(f"Hata: {e}")

    return text,res_list
```

```python
tweet_content_list = []
info_list=[]
```
