```python
def get_link_creator(url):
  html = requests.get(url).content                                                             # no filter
  soup = BeautifulSoup(html,"html.parser")
  list = soup.find_all("div",{"class":"col-12 col-sm-8 p-3"})
  test = []
  array = []
  for i in list :
    href = i.find("a").get("href")
    href = "https://bgnes.bg"+href
    with open("bgnes-Links.txt","a",encoding="utf-8") as file :
      file.write(href+"  \n")
```

```python
def creator(url):
  html = requests.get(url).content                                                                   
  soup = BeautifulSoup(html,"html.parser")

  title  = soup.find("h1",{"class":"headline"}).getText()  # Çalışıyor
  
  date = soup.find("div",{"class":"news_metadata mb-4"}).find("div").getText().strip()
  date = date[:10]
  
  txt = []
  content = ""
  cont = soup.find_all("div",{"class":"body_matrix__inner"})
  for i in cont :
    txt = i.find_all("p")
  for i in txt :
      i=i.getText().replace(" "," ")
      z=" ".join(i.split())
      content += z
  
  categori = soup.find("div",{"class":"news_categories__list mb-1"}).find("a").getText() # Çalışıyor
  sentence = "{} ; {} ; {} ; {} ; {}".format(url,date,categori,title,content)
  with open("bgnes-Cont.txt","a",encoding="utf-8") as file:
    file.write(sentence+"\n")
```

```python
import concurrent
t1=time.time()
with concurrent.futures.ProcessPoolExecutor() as execut:
  b_res=[execut.submit(creator,i.strip()) for i in total_link]
print(time.time()-t1)
```
