```python
class Daily_Hurriyet:
  technology_page_link = []
  science_page_link = []
  def date_creator(self,technology_page_link,science_page_link):
    for i in range(1,51):
      lnk = "https://www.hurriyetdailynews.com/search/science?p={}".format(str(i))
      science_page_link.append(lnk)
    for i in range(1,21):
      lnk = "https://www.hurriyetdailynews.com/search/technology?p={}".format(str(i))
      technology_page_link.append(lnk)
    return technology_page_link,science_page_link

  def technology_get_link_creator(self,url):
    html = requests.get(url).content      
    soup = BeautifulSoup(html,"html.parser")
    link_list = soup.find_all("div",{"class" : "col-md-4 col-sm-4"})
    for i in link_list:
      href = i.find("a").get("href")
      if href.startswith("https://www.hurriyetdailynews.com"):
        pass
      else:
        href = "https://www.hurriyetdailynews.com"+href
        with open("daily_hürriyet_technology_Links.txt","a",encoding="utf-8") as file :
          file.write(href+"\n")

  def science_get_link_creator(self,url):
    html = requests.get(url).content 
    soup = BeautifulSoup(html,"html.parser")
    link_list = soup.find_all("div",{"class" : "col-md-4 col-sm-4"})
    for i in link_list:
      href = i.find("a").get("href")
      if href.startswith("https://www.hurriyetdailynews.com"):
        pass
      else:
        href = "https://www.hurriyetdailynews.com"+href
        with open("daily_hürriyet_science_Links.txt","a",encoding="utf-8") as file :
          file.write(href+"\n")

  def transfer_get_link(self,HurriyetTech,HurriyetScience):
    with open("/content/daily_hürriyet_science_Links.txt",'r',newline='') as f:
      for i in f.readlines():    
        HurriyetScience.append(i)

    with open("/content/daily_hürriyet_technology_Links.txt",'r',newline='') as f:
      for i in f.readlines():    
        HurriyetTech.append(i)

    return HurriyetTech,HurriyetScience

  def creator_tech(self,url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1").getText()
    content_array = soup.find("div", attrs = {"class":"content"}).getText()
    category = "tech"
    date = soup.find("li", attrs = {"class":"date"}).getText()
    date = date.split()
    date = date[0] +"-"+ date[1] +"-"+ date[2]
    date = datetime.datetime.strptime(date,"%B-%d-%Y").strftime("%Y-%m-%d")
    content_array = content_array.split()
    content_string = ""
    for w in content_array:
      content_string = content_string +" " + w
    w_data="{};{};{};{};{}".format(url,date,category,title,content_string)
    with open("contentAll2.txt", 'a') as file: 
        file.write(w_data+'\n')

  def creator_science(self,url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1").getText()
    content_array = soup.find("div", attrs = {"class":"content"}).getText()
    category = "science"
    date = soup.find("li", attrs = {"class":"date"}).getText()
    date = date.split()
    date = date[0] +"-"+ date[1] +"-"+ date[2]
    date = datetime.datetime.strptime(date,"%B-%d-%Y").strftime("%Y-%m-%d")
    content_array = content_array.split()
    content_string = ""
    for w in content_array:
      content_string = content_string +" " + w
    w_data="{};{};{};{};{}".format(url,date,category,title,content_string)
    with open("contentAll2.txt", 'a') as file: 
        file.write(w_data+'\n')

  def daily_hurriyet_main(self):
    daily_hurriyet = Daily_Hurriyet()
    d1 = []
    d2 = []
    d3 = []
    d4 = []

    technology_page_link,science_page_link =  daily_hurriyet.date_creator(d1,d2)

    t1=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(daily_hurriyet.science_get_link_creator,i.strip()) for i in science_page_link]
    print(time.time()-t1)

    t2=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(daily_hurriyet.technology_get_link_creator,i.strip()) for i in technology_page_link]
    print(time.time()-t2)

    HurriyetTech,HurriyetScience = daily_hurriyet.transfer_get_link(d3,d4)

    t3=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(daily_hurriyet.creator_science,i.strip()) for i in HurriyetScience]
    print(time.time()-t3)

    t4=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(daily_hurriyet.creator_tech,i.strip()) for i in HurriyetTech]

    print(time.time()-t4)
```

```python
class Sputnik:
  science = []
  tech = []
  def dateCreator(self,science,tech):
    #https://sputniknews.com/services/search/getmore/?query=science&tags_limit=20&tags=0&interval=period&date_from=2021-12-01&date_to=2021-12-01   2005-02-01
    ser_date = pd.Series(pd.date_range('20050201', periods=6180))
    linkscience = "https://sputniknews.com/services/search/getmore/?query=science&tags_limit=20&tags=0&interval=period&date_from="
    linktech = "https://sputniknews.com/services/search/getmore/?query=tech&tags_limit=20&tags=0&interval=period&date_from="
    link2 = "&date_to="
    a = 0
    b = 6179  #6179 son 0 dan başlicak
    for j in range(a,b):
      dateEnd  = ser_date[j].strftime("%Y-%m-%d")
      science.append("{}{}{}{}".format(linkscience,dateEnd,link2,dateEnd))

    for j in range(a,b):#6179 son 0 dan başlicak
      dateEnd  = ser_date[j].strftime("%Y-%m-%d")
      tech.append("{}{}{}{}".format(linktech,dateEnd,link2,dateEnd))
      
    return science,tech

  def get_link_tech(self,i):
     count = 0
     r = requests.get(i)   
     soup = BeautifulSoup(r.content, 'html5lib')
     for a in soup.find_all('a', href=True):
       count+=1
       link = a['href']
       result = link.endswith("ml")
       if(result == True):
         if(count % 2 == 0):
           url  =   "https://sputniknews.com/"
           linkson = "{}{}".format(url,link)
           with open("TechLink.txt", 'a') as file: 
               file.write(linkson+'\n')

  def get_link_science(self,i):
     count = 0
     r = requests.get(i)   
     soup = BeautifulSoup(r.content, 'html5lib')
     for a in soup.find_all('a', href=True):
       count+=1
       link = a['href']
       result = link.endswith("ml")
       if(result == True):
         if(count % 2 == 0):
           url  =   "https://sputniknews.com/"
           linkson = "{}{}".format(url,link)
           with open("ScienceLink.txt", 'a') as file: 
               file.write(linkson+'\n')

  def transfer_get_link(self,SputnikNewsScience,SputnikNewsTech):
    with open("ScienceLink.txt",'r',newline='') as f:
      for i in f.readlines():    
        SputnikNewsScience.append(i)
    with open("TechLink.txt",'r',newline='') as f:
      for i in f.readlines():    
        SputnikNewsTech.append(i)
    return SputnikNewsScience,SputnikNewsTech

  def creator_tech(self,url):
      r = requests.get(url)   
      soup = BeautifulSoup(r.content, 'html5lib')
      title = soup.find("h1").getText()
      content = ""
      for wrapper in soup.find_all('div', {"data-type":"text"}):
        content = content + wrapper.text
      category = "tech"
      date = soup.find("a", attrs = {"class":"convert-date"}).getText()
      date = date.split()
      date =  date[2]
      w_data="{};{};{};{};{}".format(url,date,category,title,content)
      with open("Creator2.txt", 'a') as file: 
        file.write(w_data+'\n')
  def creator_science(self,url):
      r = requests.get(url)   
      soup = BeautifulSoup(r.content, 'html5lib')
      title = soup.find("h1").getText()
      content = ""
      for wrapper in soup.find_all('div', {"data-type":"text"}):
        content = content + wrapper.text
      category = "science"
      date = soup.find("a", attrs = {"class":"convert-date"}).getText()
      date = date.split()
      date =  date[2]
      w_data="{};{};{};{};{}".format(url,date,category,title,content)
      with open("Creator2.txt", 'a') as file: 
          file.write(w_data+'\n')
  def sputnik_main(self):
    sputnik = Sputnik()
    science = []
    tech = []
    SputnikNewsScience=[]
    SputnikNewsTech = []

    science,tech = sputnik.dateCreator(science,tech)
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(sputnik.get_link_science,i.strip()) for i in science]

    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(sputnik.get_link_tech,i.strip()) for i in tech]

    SputnikNewsScience,SputnikNewsTech = sputnik.transfer_get_link(SputnikNewsScience,SputnikNewsTech)

    t1=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(sputnik.creator_science,i.strip()) for i in SputnikNewsScience]
    print(time.time()-t1)

    t2=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(sputnik.creator_tech,i.strip()) for i in SputnikNewsTech]
    print(time.time()-t2)
```

```python
class DW_News:
  links = []  

  def dateCreator(self,links):
    ser_date = pd.Series(pd.date_range('20011001', periods=8400)) #01.10.2001 bu
    link1 = "https://www.dw.com/search/?languageCode=en&item="
    link2 = "&searchNavigationId=9097&from="
    link3 = "&to="
    link4 = "&sort=DATE&resultsCounter=50"
    category  = ['science','technology']
    a=5000
    b=5100  #7407 son
    for i in category:
      for j in range(a,b):
        dateEnd  = ser_date[j].strftime("%d.%m.%Y")
        links.append("{}{}{}{}{}{}{}".format(link1,i,link2,dateEnd,link3,dateEnd,link4))
    return links

  def get_link(self,i):
     r = requests.get(i)   
     soup = BeautifulSoup(r.content, 'html5lib')
     for a in soup.find_all('a', href=True):
       link = a['href']
       result = link.startswith("/en")
       if(result == True):
         control = link[len(link)-11:len(link)-9]
         if(control == "av" or control == "/a"):
           result2 = link.startswith("/en/travel/") or link.startswith("/en/european-union-general-data-protection-regulationgdpr-valid-may-25-2018/a-18265246") or link.startswith("/en/accessibility-statement/a-54925999")
           if(result2 == False ):
             w_data = "{}{}".format("https://www.dw.com",link)
             with open("NewsLinks.txt", 'a') as file: 
               file.write(w_data +' \n')

  def creator(self,url):
    r = requests.get(url)   
    soup = BeautifulSoup(r.content, 'html5lib')
    title = soup.find("h1").getText()
    content_array = soup.find("div", attrs = {"class":"longText"})
    content_array2 = ""
    for a in content_array.find_all('p'):
      content_array2 = content_array2 + a.getText()
    category = soup.find("h4", attrs = {"class":"artikel"}).getText()
    date = soup.find("div", attrs = {"class":"dim"}).getText()
    date = re.sub("^\s+|\s+$", "", date, flags=re.UNICODE)
    date = date.split("\n")
    date = date[1]
    content_string = ""
    content_array2 = content_array2.split(" ")
    for m in content_array2:
       content_string = content_string+" "+m
    w_data="{};{};{};{};{}".format(url,date,category,title,content_string)
    with open("contentAll.txt", 'a') as file: 
      file.write(w_data+'\n')

  def transfer_get_link(self,NewsLinksAll):
    with open("NewsLinks.txt",'r',newline='') as f:
      for i in f.readlines():    
        NewsLinksAll.append(i)
    return NewsLinksAll
  
  def dw_main(self):
    dw = DW_News()
    links = []
    NewsLinksAll = []
    links = dw.dateCreator(links)
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(dw.get_link,i.strip()) for i in links]

    NewsLinksAll = dw.transfer_get_link(NewsLinksAll)

    t1=time.time()
    with concurrent.futures.ProcessPoolExecutor() as execut:
      b_res=[execut.submit(dw.creator,i.strip()) for i in NewsLinksAll]
    print(time.time()-t1)

```
